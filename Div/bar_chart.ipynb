```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n_obs_steps = [2, 8, 12, 14, 16, 24, 32]
transformer_scores = [0.46, 0.5, 0.6, 0.62, 0.7, 0.84, 1.0]
unet_scores = [0.42, 0.54, 0.82, 0.84, 0.96, 1.0, 1.0]
```

```python
bar_width = 0.35
x = np.arange(len(n_obs_steps))

fig, ax = plt.subplots(figsize=(8, 6))

bars1 = ax.bar(x - bar_width/2, transformer_scores, bar_width, label='Transformer')
bars2 = ax.bar(x + bar_width/2, unet_scores, bar_width, label='UNet')

ax.set_xlabel('Observation Steps', fontsize=12)
ax.set_ylabel('Scores', fontsize=12)
ax.set_title('Performance Comparison by Observation Horizon', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(n_obs_steps)
ax.legend()

for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}', 
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.savefig(f"images/bar_chart_obs_horison.png", dpi=300)
plt.show()
```
