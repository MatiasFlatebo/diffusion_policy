```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
env = ['Env 1, gmts=10', 'Env 1, gmts=10', 'Env 2, gmts=10', 'Env 2, gmts=20']
unet_scores = [0.74, 0.82, 0.50, 0.46]
transformer_scores = [0.62, 0.66, 0.44, 0.48]
```

```python
bar_width = 0.35
x = np.arange(len(env))

fig, ax = plt.subplots(figsize=(10, 6))

bars1 = ax.bar(x - bar_width/2, transformer_scores, bar_width, label='Transformer')
bars2 = ax.bar(x + bar_width/2, unet_scores, bar_width, label='UNet')

ax.set_xlabel('Environment and Goal Masking Timestep', fontsize=12)
ax.set_ylabel('Scores', fontsize=12)
ax.set_title('Performance Comparison of Architectures with Image-Based Inputs', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(env, rotation=15)
ax.legend()

for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}', 
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.savefig("images/bar_chart_image.png", dpi=300)
plt.show()
```
